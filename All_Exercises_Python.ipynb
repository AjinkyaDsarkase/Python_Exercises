```python
#Exercise-4
import random
import string
st = input("Enter Message: ")
words = st.split(" ")
coding = input("1 for coding and 0 for decoding")
coding = True if (coding=='1') else False
if(coding):
  nwords = []
  for word in words:
    if(len(word)>=3):
      r1 = "".join(random.choices(string.ascii_letters, k = 3))
      r2 = "".join(random.choices(string.ascii_letters, k = 3))
      stnew = r1 + word[1:] + word[0] + r2
      nwords.append(stnew)
    else:
      nwords.append(word[::-1])
  print(" ".join(nwords))
else:
  nwords = []
  for word in words:
    if(len(word)>=3):
      stnew = word[3:-3]
      stnew = stnew[-1] + stnew[:-1]
      nwords.append(stnew)
    else:
      nwords.append(word[::-1])
  print(" ".join(nwords))
```

```python
#Exercise-5(Snake, Water, Gun)

import random
print("Snake Water Gun Game")
l = [0, 1, 2]
print("0 = Snake")
print("1 = Water")
print("2 = Gun")

comp = random.choice(l)
a  = int(input("Enter your choice: "))
print("Randomly selected Computer's choice was: ",comp)
if a == comp:
  print("Draw")
elif a == 0 and comp == 1:
  print("You Won")
elif a == 0 and comp == 2:
  print("You loose")

elif a == 1 and comp == 2:
  print("You Won")
elif a == 1 and comp == 0:
  print("You loose")

elif a == 2 and comp == 0:
  print("You Won")
elif a == 2 and comp == 1:
  print("You loose")
else:
  print("Wrong input")
```
